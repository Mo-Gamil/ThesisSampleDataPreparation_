```python
import geemap
import ee
import os
import datetime

```

```python
# Set up Earth Engine credentials
ee.Authenticate()
ee.Initialize()
```

```python


# Define the region of interest from the shapefile
shapefile_path = "grid_geometries/grid_geometries_Iowa_Project.shp"
roi = geemap.shp_to_ee(shapefile_path)

# Define the time range (July 2021)
start_date = '2021-07-01'
end_date = '2021-07-31'
time_range = (ee.Date(start_date), ee.Date(end_date))

# Define the Sentinel-2 bands to be downloaded
bands = ['B2', 'B3', 'B4', 'B5']

# Cloud cover threshold
cloud_cover_max = 11

# Loop through each feature in the shapefile
fc = geemap.shp_to_ee(shapefile_path)
features = fc.getInfo()['features']
for feature in features:
    feature_name = feature['properties']['Name']  

    # Construct the image collection
    collection = (ee.ImageCollection('COPERNICUS/S2')
                  .filterBounds(roi)
                  .filterDate(time_range[0], time_range[1])
                  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', cloud_cover_max)))

    # Calculate mean reflectance
    mean_image = collection.mean()

    # Select specific bands
    selected_image = mean_image.select(bands)

    # Get the geometry of the feature
    geometry = ee.Feature(feature['geometry']).geometry()

    # Download the image to local drive
    output_path = os.path.join("sen2Img", f"{feature_name}_Iowa_2021.tif")
    geemap.ee_export_image(selected_image, filename=output_path, scale=10, region=geometry)

    print(f"Downloaded {feature_name} image patch for Iowa 2021.")

print("Download completed.")
```

```python
import os

def count_files_in_folder(folder_path):
    try:
        # Get the list of files in the folder
        files = os.listdir(folder_path)
        
        # Filter out only files (excluding directories)
        files = [f for f in files if os.path.isfile(os.path.join(folder_path, f))]
        
        # Return the count of files
        return len(files)
    
    except FileNotFoundError:
        print(f"The folder '{folder_path}' does not exist.")
        return -1

# Example usage:
folder_path = "sen2Img"
file_count = count_files_in_folder(folder_path)

if file_count >= 0:
    print(f"The number of files in '{folder_path}' is: {file_count}")
```
